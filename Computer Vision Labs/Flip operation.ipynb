```python
import cv2
from google.colab.patches import cv2_imshow # Import the cv2_imshow function

img=cv2.imread('/content/approach category.png')
cv2_imshow(img) # Use cv2_imshow instead of cv2.imshow
#cv2.waitkey() # cv2.waitKey is not needed with cv2_imshow in Colab
```

```python
img1=cv2.resize(img,(400,400))
cv2_imshow(img1)
```

```python
gray_img=cv2.imread('/content/approach category.png',0)
cv2_imshow(gray_img)
```

```python
#increase channnel values using alpha to show saturation
gray_img=cv2.imread('/content/approach category.png',-1)
cv2_imshow(gray_img)
```

```python
img=cv2.imread('/content/approach category.png',0)
k=cv2.waitKey()
if k==ord('s'):# when ever s print then img of gray save by pressing s on keyboard
  cv2.imwrite('/content/approach_gray.png',img)
else:
  cv2.destroyAllWindows()
```
