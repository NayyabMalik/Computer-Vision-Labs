## image histogram
### x represent range of color values
### y represent number of pixels values
### function:calcHist(img,channel,mask,histsize,range(0,255))

```python
img=np.zeros((200,200),np.uint8)
hist=cv2.calcHist([img],[0],None,[256],[0,255])
plt.plot(hist)
plt.show()
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Color image histogram

```python
img=cv2.imread(r"C:\Users\PMLS\Documents\computer vision\panda.jpeg")
img=cv2.resize(img,(500,400))
b,g,r=cv2.split(img)

plt.hist(b.ravel(),256,[0,256])
plt.hist(g.ravel(),256,[0,256])
plt.hist(r.ravel(),256,[0,256])

plt.title('colorful images')
plt.show()
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# gray scale image histogram

```python
img=cv2.imread(r"C:\Users\PMLS\Documents\computer vision\panda.jpeg")
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray_img=cv2.resize(gray_img,(500,400))
hist=cv2.calcHist([gray_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.title('gray images')
plt.show()
cv2.imshow('img',gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Equalize histogram :gray scale image accept


```python
equ=cv2.equalizeHist(gray_img)
res=np.hstack((gray_img,equ))
cv2.imshow('equ',res)
hist=cv2.calcHist([equ],[0],None,[256],[0,256])
plt.plot(hist)
plt.title('gray images')
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Contrast limited adaptive histogram equalization CLAHE 

```python
img=cv2.imread(r"C:\Users\PMLS\Documents\computer vision\panda.jpeg")
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray_img=cv2.resize(gray_img,(500,400))
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))## CLAHE remove noise from images
cl1=clahe.apply(gray_img)
hist=cv2.calcHist([cl1],[0],None,[256],[0,256])
plt.plot(hist)
plt.title('CLAHE ')
plt.show()
cv2.imshow('CLAHE',cl1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
