```python
import numpy as np
import cv2
```

## image contours are used to detect,analyze and recognize the shapes from images
## contours are the curve joining  all continuous points depending on color and intensity 

```python
img=cv2.imread(r"C:\Users\PMLS\Documents\computer vision\panda.jpeg")
img=cv2.resize(img,(600,400))
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


## It is necessary for contour to convert the background into black and the front to a white 
_,mask=cv2.threshold(img_gray,12,255,1)
smoothing=cv2.medianBlur(mask,3)
canny=cv2.Canny(smoothing,100,200)


## contour, hierarchy=findcountours(thresh,cv3,Retrieval mode ,method )
cnt,hiers=cv2.findContours(canny,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
print(cnt,len(cnt))


##draw contour to see results that we want to detect 
## cv2.drawContours(img,contour,id of contours,color,thickness)
img_draw=cv2.drawContours(img,cnt,-1,(170,50,30),4)
# cv2.imshow('img',img)
# cv2.imshow('img_gray',img_gray)
# cv2.imshow('mask',mask)
# cv2.imshow('smoothing',smoothing)
# cv2.imshow('canny',canny)
cv2.imshow('drawContours',img_draw)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## to choose specific contour 
img_draw=cv2.drawContours(img,cnt,5,(170,50,30),4)
cv2.imshow('draw contours',img_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
